```python
data.head()
data['road']=0
data['city']=0
data['town']=0
data['state']=0

data.head()
```

```python
data
n=-1
for i in data.values:
    n=n+1
    locator = Nominatim(user_agent='myGeocoder')
    latitude=str(i[1])
    longitude=str(i[2])
    coordinates=latitude+', '+longitude
    location = locator.reverse(coordinates)
    while True:
        try:
            print(location.raw['address']['road'],',',location.raw['address']['city'],',',location.raw['address']['state'])
            data['road'][n]=location.raw['address']['road']
            data['city'][n]=location.raw['address']['city']
            data['state'][n]=location.raw['address']['state']
            break
        except KeyError:
            try:
                print('0,',location.raw['address']['city'],',',location.raw['address']['state'])
                data['city'][n]=location.raw['address']['city']
                data['state'][n]=location.raw['address']['state']
                break
            except KeyError:
                try:
                    print('0,', location.raw['address']['town'],',',location.raw['address']['state'])
                    data['town'][n]= location.raw['address']['town']
                    data['state'][n]=location.raw['address']['state']
                    break
                except KeyError:
                    try:
                        print('0,', '0,',location.raw['address']['state'])
                        data['state'][n]=location.raw['address']['state']
                        break
                    except KeyError:
                        break

#print(location.raw['address']['road'],',',location.raw['address']['city'],',',location.raw['address']['state'])#',',location.raw['address']['city'],
```
